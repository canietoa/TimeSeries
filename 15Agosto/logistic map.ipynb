```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def sn(s,n):
    return (s+n)
```

```python
narray=[]
snarray=[]
s0=0
for n in range(10):
    s0 = s0+n#sn(s0,n)
    narray.append(n)
    snarray.append(s0)
```

```python
plt.scatter(narray,snarray)
plt.grid()
```

```python
limarr=[]
rarr=[]
for r in np.linspace(0.1,4,100):
    x=0.5#condicion inicial
    for n in range(200):
        x=r*x*(1-x)
        if n>150:
            limarr.append(x)
            rarr.append(r)
    plt.scatter(rarr,limarr,c='g',s=1)    
```
