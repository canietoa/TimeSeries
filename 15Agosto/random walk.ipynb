```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#llamamos 0 a salir cara
#si sale cara (0), le asignamos 1 a ci
#si sale sello (1) le asignamos -1 a ci 
def ci(i):
    if i==0:
        return 1
    else:
        return -1   
```

```python
#generamos 1000 numeros aleatorios
#para cada numero aleatorio calculamos ci
#hacemos una sucesion de los ci y le calculamos el promedio
#verificamos que sea 0
dataarr=[]
for n in range(10000):
    x=np.random.randint(2)#generamos un numero aleatorio entre 0 y 1 (entero)
    c=ci(x)
    dataarr.append(c)#guardando los numero aleatorios
print(np.mean(dataarr))
print(np.var(dataarr))
```

```python
#para cada instante de tiempo lanzo una moneda y se la adiciono al valor actual
dataarr=[]
sn=0#condicion inicial
for n in range (10):#diez instantes de tiempo
    dataarr.append(sn)
    cn=ci(np.random.randint(2))#lanzo la moneda
    sn=sn+cn
plt.plot(dataarr)   
```

```python
#serie estocastica

for l in range (1000):#graficar 20 caminatas aleatorias
    #para cada instante de tiempo lanzo una moneda y se la adiciono al valor actual
    dataarr=[]
    sn=0#condicion inicial
    for n in range (50):#cincuenta instantes de tiempo
        dataarr.append(sn)
        cn=ci(np.random.randint(2))#lanzo la moneda
        sn=sn+cn
    plt.plot(dataarr)
plt.grid()
```

```python
#Como cambia el promedio de la serie a lo largo del tiempo
#guardamos los valores del ultimo instante y le calculamos el promedio y la varianza
sNarr=[] #los datos de cada serie en el ultimo instante
for l in range (1000):#graficar 20 caminatas aleatorias
    #para cada instante de tiempo lanzo una moneda y se la adiciono al valor actual
    dataarr=[]
    sn=0#condicion inicial
    for n in range (50):#cincuenta instantes de tiempo
        dataarr.append(sn)
        cn=ci(np.random.randint(2))#lanzo la moneda
        sn=sn+cn#a la serie le sumo el valor que asume la mones
    plt.plot(dataarr)#grafico la caminata
    sNarr.append(sn)#guardo el ultimo valor (sn) de la serie en el areglo de valores de sN
```

```python
print(np.mean(sNarr))#promedio de la serie hasta 50
```

```python
print(np.var(sNarr))#varianza de la serie hasta 50
```

```python
plt.hist(sNarr,bins=50)
```

```python
#Que tanto se puede predecir el valor futuro conociendo el valor presente?
#Como cambia el promedio de la serie a lo largo del tiempo
#guardamos los valores del ultimo instante y le calculamos el promedio y la varianza
sNarr=[] #los datos de cada serie en el ultimo instante
sNarr1=[] #los datos de cada serie en el penultimo instante
for l in range (1000):#graficar 1000 caminatas aleatorias
    #para cada instante de tiempo lanzo una moneda y se la adiciono al valor actual
    dataarr=[]
    sn=0#condicion inicial
    for n in range (50):#cincuenta instantes de tiempo
        dataarr.append(sn)
        cn=ci(np.random.randint(2))#lanzo la moneda
        sn=sn+cn#a la serie le sumo el valor que asume la mones
        if n==49:
            ultimodato=sn
        if n==10:
            penultimodato=sn
    plt.plot(dataarr)#grafico la caminata
    sNarr.append(ultimodato)#guardo el ultimo valor (sn) de la serie en el areglo de valores de sN
    sNarr1.append(penultimodato)#guardo el ultimo valor (sn) de la serie en el areglo de valores de sN
```

```python
plt.scatter(sNarr1,sNarr)
#coefficiente de correlacion
print(np.corrcoef(sNarr1,sNarr)[0][1])

#autocorrelacion entre el instante 10 y el instante 50
```

```python
#la correlacion es 0 cuando no hay relacion
#es 1 cuando hay correlacion perfecta
#Que tanto se puede predecir el valor futuro conociendo el valor presente?
#Como cambia el promedio de la serie a lo largo del tiempo
#guardamos los valores del ultimo instante y le calculamos el promedio y la varianza
corrarr=[]

for t in range(49):
    sNarr=[] #los datos de cada serie en el ultimo instante
    sNarr1=[] #los datos de cada serie en el instante t
    for l in range (1000):#graficar 1000 caminatas aleatorias
        #para cada instante de tiempo lanzo una moneda y se la adiciono al valor actual
        dataarr=[]
        sn=0#condicion inicial
        for n in range (50):#cincuenta instantes de tiempo
            dataarr.append(sn)
            cn=ci(np.random.randint(2))#lanzo la moneda
            sn=sn+cn#a la serie le sumo el valor que asume la moneda
            if n==49:
                ultimodato=sn
            if n==49-t:
                penultimodato=sn
        #plt.plot(dataarr)#grafico la caminata
        sNarr.append(ultimodato)#guardo el ultimo valor (sn) de la serie en el areglo de valores de sN
        sNarr1.append(penultimodato)#guardo el valor de sn de la serie en el tiempo 49-t
    corrarr.append(np.corrcoef(sNarr,sNarr1)[0][1])
```

```python
plt.plot(corrarr)
```
