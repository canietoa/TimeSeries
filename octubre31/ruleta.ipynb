```python
def ruleta():#1 si es rojo y 0 si es negro
    return random.randint(0,1)
def apuesta():#defino apuesta aleatoria
    return random.randint(0,1)
def gane(ruleta,apuesta):#si apuesta es igual a ruleta, gano, sino, pierdo
    if ruleta==apuesta:
        return True
    else:
        return False
def pago(dinero,resultado):
    if resultado:
        return dinero+10000
    else:
        return dinero-10000
```

```python
#una sola trayectoria
intentos = 100
dinero=100000
dinarr=np.zeros(intentos)#dinero a lo largo de los intentos

for m in range(intentos):
    apt=apuesta()
    rlt=ruleta()
    gn=gane(rlt,apt)
    dinero=pago(dinero,gn)
    dinarr[m]=dinero
    if dinero<10000:
        break
```

```python
plt.plot(dinarr)
```
