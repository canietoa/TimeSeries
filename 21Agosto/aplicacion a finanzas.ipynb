```python
nombres=data.Name.unique()#obtener todos los valores de los nombres
for name in nombres:#iterar sobre los nombres
    df=data[data.Name==name]#Escoger dataos con el nombre name
    df['Date']=pd.to_datetime(df.Date, format='%m/%d/%Y') #convertir los datos en Date a "tipo" fecha
    df=df.sort_values(by='Date',ascending=True)#ordenar los datos
    x=np.arange(len(df))#arreglo de dias de longitud igual a df
    s0=df.Close.tolist()[0]#obteniendo el primer dato (convierto los datos a lista y escojo el primer valor de la lista)
    plt.plot(x,df.Close/s0)#graficar normalizado al primer valor
    plt.text(x[-1],df.Close.tolist()[-1]/s0,name)#escribir los nombres de cada accion al final de la serie de tiempo
plt.grid()
```

```python
from sklearn.linear_model import LinearRegression
palette = plt.get_cmap('tab20')
num=0
nombres=data.Name.unique()#obtener todos los valores de los nombres
for name in nombres:#iterar sobre los nombres
    df=data[data.Name==name]#Escoger dataos con el nombre name
    df['Date']=pd.to_datetime(df.Date, format='%m/%d/%Y') #convertir los datos en Date a "tipo" fecha
    df=df.sort_values(by='Date',ascending=True)#ordenar los datos
    x=np.arange(len(df))#arreglo de dias de longitud igual a df
    s0=df.Close.tolist()[0]#obteniendo el primer dato (convierto los datos a lista y escojo el primer valor de la lista)
    s=np.array(df.Close)
    y=np.array(np.log(s/s0))
    
    model = LinearRegression(fit_intercept=False)
    model.fit(x.reshape((-1, 1)), y)#defino la ajuste (fit)
    mu0=model.coef_#Regresa la pendiente de la regresion
    print(name)
    print(mu0)
    plt.plot(x,np.exp(mu0[0]*x),color=palette(num))
    plt.plot(x,df.Close/s0,color=palette(num))#graficar normalizado al primer valor
    plt.text(x[-1],df.Close.tolist()[-1]/s0,name)#escribir los nombres de cada accion al final de la serie de tiempo
    num+=1
```
