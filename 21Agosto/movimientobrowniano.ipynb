```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#para cada instante de tiempo lanzo una moneda y se la adiciono al valor actual
dataarr=[]
sn=0#condicion inicial
for n in range (10):#diez instantes de tiempo
    dataarr.append(sn)
    sn=sn+np.random.normal(0,1)
plt.plot(dataarr)   
```

```python
for l in range (100):#graficar 20 caminatas aleatorias
    #para cada instante de tiempo lanzo una moneda y se la adiciono al valor actual
    dataarr=[]
    sn=0#condicion inicial
    for n in range (50):#cincuenta instantes de tiempo
        dataarr.append(sn)
        sn=sn+np.random.normal(0,1)
    plt.plot(dataarr)
plt.grid()
```

```python
#MOVIMIENTO BROWNIANO
for l in range (100):#graficar 20 caminatas aleatorias
    #para cada instante de tiempo lanzo una moneda y se la adiciono al valor actual
    t=0
    dt=0.01
    dataarr=[]
    timearr=[]
    sn=0#condicion inicial
    for n in range (50):#cincuenta instantes de tiempo
        dataarr.append(sn)
        timearr.append(t)
        t=t+dt
        sn=sn+np.random.normal(0,np.sqrt(dt))
    plt.plot(timearr,dataarr)
plt.grid()
```

```python
t=0
y=1
mu=2
dt=0.001
tmax=3
yarr=[]
tarr=[]
while(t<tmax):
    yarr.append(y)
    tarr.append(t)
    y=y+mu*y*dt
    t=t+dt
    
```

```python
plt.plot(tarr,yarr,label="Numerical")
plt.plot(tarr,np.exp(mu*np.array(tarr)),label="Theoretical")
plt.legend()
```

```python
#OU process

t=0
y=0
mu=2
dt=0.001
tmax=3
yarr=[]
tarr=[]
while(t<tmax):
    yarr.append(y)
    tarr.append(t)
    y=y-mu*y*dt+np.random.normal(0,np.sqrt(dt))
    t=t+dt
```

```python
plt.plot(tarr,yarr)
```

```python
#OU process
for l in range (100):
    #para cada instante de tiempo lanzo una moneda y se la adiciono al valor actual
    t=0
    y=0
    mu=2
    dt=0.001
    sigma=1
    tmax=3
    yarr=[]
    tarr=[]
    while(t<tmax):
        yarr.append(y)
        tarr.append(t)
        y=y-mu*y*dt+sigma*np.random.normal(0,np.sqrt(dt))
        t=t+dt
    plt.plot(tarr,yarr)
plt.grid()
```

```python
#MOVIMIENTO BROWNIANO GEOMETRICO (Finanzas)
for l in range (200):
    #para cada instante de tiempo lanzo una moneda y se la adiciono al valor actual
    t=0
    y=1
    mu=1.1
    dt=0.001
    sigma=1#volatilidad
    tmax=3
    yarr=[]
    tarr=[]
    count=0
    while(t<tmax):
        count+=1
        if count==10:
            yarr.append(y)
            tarr.append(t)
            count=0
        y=y+mu*y*dt+sigma*y*np.random.normal(0,np.sqrt(dt))
        t=t+dt
    plt.plot(tarr,yarr)
plt.grid()
plt.plot(tarr,np.exp(mu*np.array(tarr)),c='k')
plt.yscale("log")
```

```python
#Como cambia el promedio de la serie a lo largo del tiempo
#guardamos los valores del ultimo instante y le calculamos el promedio y la varianza
sNarr=[] #los datos de cada serie en el ultimo instante
for l in range (5000):#graficar 20 caminatas aleatorias
    #para cada instante de tiempo lanzo una moneda y se la adiciono al valor actual
    t=0
    y=1
    mu=1.1
    dt=0.001
    sigma=1#volatilidad
    tmax=3
    yarr=[]
    tarr=[]
    count=0
    while(t<tmax):
        count+=1
        if count==10:
            yarr.append(y)
            tarr.append(t)
            count=0
        y=y+mu*y*dt+sigma*y*np.random.normal(0,np.sqrt(dt))
        t=t+dt
    plt.plot(tarr,yarr)
    sNarr.append(y)#guardo el ultimo valor (sn) de la serie en el areglo de valores de sN
```
