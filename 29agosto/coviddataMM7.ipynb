```python
#seleccionar datos de "United States"
df=data[data.location=='United States']
df['date']=pd.to_datetime(df.date, format='%m/%d/%Y') #convierte los valores de "date" en datos tipo fecha
df=df.sort_values(by='date')#Ordena los datos por "date"
x=np.arange(len(df.new_deaths))
plt.plot(x,df.new_deaths)
plt.grid()
```

```python
#seleccionar datos de "United States"
df=data[data.location=='United States']
df['date']=pd.to_datetime(df.date, format='%m/%d/%Y') #convierte los valores de "date" en datos tipo fecha
df=df.sort_values(by='date')#Ordena los datos por "date"
df["MM7"]=df.new_cases.rolling(window=7).mean()#nueva columna llamada MM7 con la media movila de 7 dias en "new_cases"
x=np.arange(len(df.new_cases))
plt.plot(x,df.MM7)
plt.plot(x,df.new_cases)
plt.grid()
```
