```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
datos=pd.read_csv("./dataARMA.csv")
```

```python
df=datos["serie1"]
```

```python
import statsmodels.graphics.tsaplots as sm
```

```python
sm.plot_acf(df,lags=11)
```

```python
sm.plot_pacf(df,lags=11)
```
