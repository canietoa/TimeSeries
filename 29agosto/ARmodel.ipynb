```python
#simulando un proceso AR(1)
data=np.zeros(1000)
data[0]=np.random.normal(0,1)
#print(data[0])
for l in range(1,1000):
    data[l]=-0.5*data[l-1]+0.2*np.random.normal(0,1)#modelo tipo AR(1)
```

```python
plt.plot(data[0:30])
```

```python
import statsmodels.graphics.tsaplots as sm
sm.plot_acf(data,lags=11)
```

```python
sm.plot_pacf(data,lags=11)#partial autocorrelation
```

```python
import statsmodels.tsa.ar_model as smar
```

```python
#simulando un proceso AR(2)
data=np.zeros(1000)
data[0]=np.random.normal(0,1)
data[1]=-0.5*data[0]+0.2*np.random.normal(0,1)
for l in range(2,1000):
    data[l]=-0.7*data[l-2]-0.5*data[l-1]+0.2*np.random.normal(0,1)#modelo tipo AR(2)
```

```python
model=smar.AR(data)
model_fit = model.fit()
print('Coefficients: %s'% model_fit.params)
```
