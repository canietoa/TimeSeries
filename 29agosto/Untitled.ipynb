```python
noises=np.zeros(1000)
data=np.zeros(1000)
for l in range(len(data)):
    noises[l]=np.random.normal(0,1)
    if l == 0:
        data[l]=noises[l]
    else:
        data[l]=0.5*data[l-1]+noises[l]
```

```python
import statsmodels.tsa.stattools as sm
v,confint=sm.acf(data, unbiased=False, nlags=10, qstat=False, fft=None, alpha=0.05, missing='none')
plt.scatter(range(len(v)),v)
plt.plot(v-confint[:,1])
plt.plot(v-confint[:,0])
```

```python
import statsmodels.tsa.stattools as sm
v,confint=sm.pacf(data, nlags=10, method='ywunbiased', alpha=0.05)
#plt.errorbar(range(len(v)),v,confint[:,1],capsize=6,markersize='10',elinewidth=3,lw=3)
plt.scatter(range(len(v)),v)
plt.plot(v-confint[:,1])
plt.plot(v-confint[:,0])
#plt.plot(confint[:,1])
#plt.plot(confint[:,0])
```

```python
import statsmodels.tsa.ar_model as sm
df=data[:-5]
```

```python
import statsmodels.tsa.ar_model as sm
df=data
model=sm.AR(df)
model_fit = model.fit()
print('Coefficients: %s' % model_fit.params)

```

```python
pred=model_fit.predict(start=1000, end=1100, dynamic=False)
```
