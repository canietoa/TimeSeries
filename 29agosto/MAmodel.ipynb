```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#simulando un proceso MA(1)
data=np.zeros(1000)
noises=np.random.normal(0,1,1000)
data[0]=np.random.normal(0,1)
for l in range(1,1000):
    data[l]=+0.8*noises[l-1]+noises[l]#modelo tipo MA(1)
```

```python
plt.plot(data[0:30])
plt.plot(noises[0:30])
```

```python
import statsmodels.graphics.tsaplots as sm
```

```python
sm.plot_acf(data,lags=11)
```

```python
sm.plot_pacf(data,lags=11)
```

```python
#simulando un proceso ARMA(1,1)
data=np.zeros(1000)
noises=np.random.normal(0,1,1000)
data[0]=noises[0]
for l in range(1,1000):
    data[l]=0.2*data[l-1]+0.7*noises[l-1]+noises[l]
```

```python
sm.plot_acf(data,lags=11)
```

```python
sm.plot_pacf(data,lags=11)
```

```python
#simulando un proceso ARMA(1,1)
data=np.zeros(10000)
noises=np.random.normal(0,1,10000)
data[0]=noises[0]
for l in range(1,10000):
    data[l]=0.2*data[l-1]+0.7*noises[l-1]+noises[l]
```

```python
import statsmodels.tsa.arima_model as smarma
```

```python
model55=smarma.ARMA(data,order=(1,1))
model_fit = model55.fit()
print('Coefficients: %s' % model_fit.params)
```

```python
#simulando un proceso ARMA(2,2)
data=np.zeros(10000)
noises=np.random.normal(0,1,10000)
data[0]=noises[0]
data[1]=0.2*data[0]+0.7*noises[0]+noises[1]
for l in range(2,10000):
    data[l]=0.6*data[l-2]-0.7*noises[l-2]+0.2*data[l-1]+0.7*noises[l-1]+noises[l]
```

```python
model55=smarma.ARMA(data,order=(2,2))
model_fit = model55.fit()
print('Coefficients: %s' % model_fit.params)
```
