```python
data=pd.read_csv("./DatosAutomoviles.csv",sep=";")
companies=['alfa-romero','audi','bmw','jaguar','mercedes-benz','volvo']#companias que quiero fliltrar
df=pd.DataFrame(columns=data.columns)#dataframe vacio con columnas iguales a data
for comp in companies:#iterando ent las companias que he elegido
    dftemp=data[data['company']==comp]#haciendo un nuevo dataframes con la compania que estoy iterando
    df=pd.concat([df,dftemp])#concatenando dataframes
    
```

```python
#Hacer una grafica de scatter entre horsepower y price

plt.scatter(data.horsepower,data.price,c="#FE5E48",s=20,label="Todo el mundo")
plt.scatter(df.horsepower,df.price,c="#23BD00",s=10,label="Marcas Europeas")
plt.legend(loc="lower right",fontsize=15)
plt.grid()
plt.xlabel("Price",fontsize=15)
plt.ylabel("Horse Power",fontsize=15)
#for axis in ['top','bottom','left','right']:
#    ax.spines[axis].set_linewidth(2)
    #plt.tick_params(axis='both', width=2,length=6)
```
